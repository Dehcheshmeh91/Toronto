Using Pandas the following table was created.

```python
dfs = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',header=0)
df = dfs[0]
df.head()
```

All the "Not assigned" values were deleted using the following code.

```python
df = df[df.Borough != 'Not assigned']
```

```python
df
```

Using the following groupby we could combine the rows. 

```python
df = df.groupby(by=['Postcode','Borough']).agg(lambda x: ','.join(x))
```

```python
df = df.replace({'Not assigned' : "Queen's Park"}) # One Not assigned value was replaced in the Neighboirhood column
df.head()
```

```python
df = df.rename(columns={'Postcode' : 'Postal Code'}) # The column name was changed.
```

```python
df = df.reset_index()
```

```python
df.head()
```
